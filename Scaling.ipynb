```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/tkseneee/Dataset/master/k_circle_sales.csv')
```

```python
num_data=data.select_dtypes(include=np.number)
```

```python
num_data.isnull().sum()
```

```python
num_data['Item_Weight']=num_data['Item_Weight'].fillna(num_data['Item_Weight'].mean())
```

```python
num_data.isnull().sum()
```

```python
#Range of values
num_data.max()-num_data.min()
```

Scalar Types
 - Standard Scalar
 - MinMax Scalar
 - Robust Scalar

 Standard Scalar ( Standardization)


```python
num_data_sc=num_data
num_data_sc['Item_Outlet_Sales']=(num_data['Item_Outlet_Sales']-num_data['Item_Outlet_Sales']
                                   .mean())/num_data['Item_Outlet_Sales'].std()
```

```python
    # standard scale to all the features
for i in num_data_sc.columns:
  num_data_sc[i]=(num_data_sc[i]-num_data_sc[i].mean())/num_data_sc[i].std()
```

```python
num_data_sc.describe()
```

```python
#Range of values
num_data.max()-num_data.min()
```

```python
num_data_mmsc=num_data
num_data.max()-num_data.min()
```

```python
num_data_mmsc['Item_Outlet_Sales']=(num_data_mmsc['Item_Outlet_Sales']-num_data_mmsc['Item_Outlet_Sales'].min())/
 (num_data_mmsc['Item_Outlet_Sales'].max()-num_data_mmsc['Item_Outlet_Sales'].min())
```

```python
num_data_mmsc['Item_Outlet_Sales'].describe()
```

```python
for i in num_data_mmsc.columns:
  num_data_mmsc[i]=(num_data_mmsc[i]-num_data_mmsc[i].mean())/(num_data_mmsc[i].max()-num_data_mmsc[i].mean())
```

```python
num_data_mmsc.describe()
```

```python
# Robust Scalar
num_data_rbsc=num_data
# X-q1/IQR
```

```python
q1=num_data_rbsc['Item_Outlet_Sales'].quantile(.25)
q3=num_data_rbsc['Item_Outlet_Sales'].quantile(.75)
IQR=q3-q1
num_data_rbsc['Item_Outlet_Sales']=
 (num_data_rbsc['Item_Outlet_Sales']-q1)/IQR
```

```python
for i in num_data_rbsc.columns:
  q1=num_data_rbsc[i].quantile(.25)
  q3=num_data_rbsc[i].quantile(.75)
  IQR=q3-q1
  num_data_rbsc['Item_Outlet_Sales']=(num_data_rbsc[i]-q1)/IQR
```

```python
for i in num_data.columns:
  print(i)
  plt.subplot(1,2,1)
  num_data[i].plot(kind='kde')
  plt.subplot(1,2,2)
  num_data_rbsc[i].plot(kind='kde')
  plt.show()
```

```python
data_sc=num_data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_sc=scaler.fit_transform(data_sc)
data_sc=pd.DataFrame(data_sc,columns=num_data.columns)
```

```python
data_mm=num_data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_mm=scaler.fit_transform(data_mm)
data_mm=pd.DataFrame(data_mm,columns=num_data.columns)
data_mm.describe()
```

```python
data_rb=num_data
from sklearn.preprocessing import RobustScaler
scaler = RoubstScaler()
data_rb=scaler.fit_transform(data_rb)
data_rb=pd.DataFrame(data_rb,columns=num_data.columns)
```

```python
# normailzer
data_nor=num_data
from sklearn.preprocessing import Normalizer
norm=Normalizer()
data_nor=norm.fit_transform(data_nor)
data_nor=pd.DataFrame(data_nor,columns=num_data.columns)
```

```python
for i in num_data.columns:
  print(i)
  plt.subplot(1,2,1)
  num_data[i].plot(kind='kde')
  plt.subplot(1,2,2)
  data_nor[i].plot(kind='kde')
  plt.show()
```
